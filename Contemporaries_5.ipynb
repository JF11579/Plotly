# Preprocessing

```python
data_people['Name'] = data_people['Name'].astype(str)
data_people['Name'] = data_people['Name'].str.strip()
data_people['Name'] = data_people['Name'].str.replace('[^a-zA-Z0-9 ]', '', regex=True)
data_people = data_people.sort_values(by=['Name'])
```

## Bow Add Histrocial events

```python
# 2

# Convert 'Born' and 'Died' columns to string format
data_people['Born'] = data_people['Born'].astype(str)
data_people['Died'] = data_people['Died'].astype(str)

# Plotly Express timeline plot with category_orders parameter
fig = px.timeline(data_people, x_start='Born', x_end='Died', y='Name', color='Field',
                  labels={'Name': 'Name'},
                  category_orders={'Name': data_people['Name'].tolist()})


###############################################
#             EVENTS                          #
###############################################

################################################################################
#                        World War One (1914-1918)
fig.add_shape(
    type='rect',
    x0='1914-01-01',
    x1='1918-12-31',
    y0=0,
    y1=len(data_people['Name']),
    fillcolor='rgba(255, 0, 0, 0.2)',
    line=dict(color='black', width=3),
    layer='below',
)

# Add a label to the vertical bar with line breaks and bold text
fig.add_annotation(
    text='<b>World<br>War<br>One',
    x='1916-06-15',  # Adjust the x-coordinate for label placement
    y=len(data_people['Name']) * -0.1,  # Adjust the y-coordinate for label placement
    showarrow=False,
    font=dict(color='white', size=14),  # Set the text size
)
################################################################################
#                    Add a vertical band for WW2

fig.add_shape(
    type='rect',
    x0='1939-01-01',
    x1='1945-12-31',
    y0=0,
    y1=len(data_people['Name']),
    fillcolor='rgba(255, 0, 0, 0.2)',
    line=dict(color='black', width=3),
    layer='below',
)

# Add a label to the vertical bar with line breaks and bold text
fig.add_annotation(
    text='<b>World<br>War<br>Two',
    x='1942-06-15',  # Adjust the x-coordinate for label placement
    y=len(data_people['Name']) * -0.1,  # Adjust the y-coordinate for label placement
    showarrow=False,
    font=dict(color='white', size=14),  # Set the text size
)

##############################################################################
##############################################################################
#                      American Civil War

fig.add_shape(
    type='rect',
    x0='1861-01-01',
    x1='1865-12-31',
    y0=0,
    y1=len(data_people['Name']),
    fillcolor='rgba(255, 0, 0, 0.2)',
    line=dict(color='black', width=3),
    layer='below',
)

# Add a label to the vertical bar with line breaks and bold text
fig.add_annotation(
    text='<b>American<br>Civil<br>War',
    x='1863-06-15',  # Adjust the x-coordinate for label placement
    y=len(data_people['Name']) *1.1,  # Adjust the y-coordinate for label placement
    showarrow=False,
    font=dict(color='white', size=14),  # Set the text size
)
##############################################################################
##############################################################################
#                        Tulsa Race Riot

fig.add_shape(
    type='rect',
    x0='1921-01-01',
    x1='1923-12-31',
    y0=0,
    y1=len(data_people['Name']),
    fillcolor='rgba(255, 0, 0, 0.2)',
    line=dict(color='black', width=3),
    layer='below',
)

# Add a label to the vertical bar with line breaks and bold text
fig.add_annotation(
    text='<b>Tulsa<br>Race<br>Riot</b>',
    x='1922-01-01',  # Adjust the x-coordinate for label placement
    y=len(data_people['Name']) * 1.1,  # Adjust the y-coordinate for label placement
    showarrow=False,
    font=dict(color='white', size=14, family='Arial, sans-serif'),  # Set the text size and family
)

##############################################################################
#              American Revolution

fig.add_shape(
    type='rect',
    x0='1775-01-01',
    x1='1783-12-31',
    y0=0,
    y1=len(data_people['Name']),
    fillcolor='rgba(255, 0, 0, 0.2)',
    line=dict(color='black', width=3),
    layer='below',
)

# Add a label to the vertical bar with line breaks and bold text
fig.add_annotation(
    text='<b>American<br>Revolution<br></b>',
    x='1779-01-01',  # Adjust the x-coordinate for label placement
    y=len(data_people['Name']) * 1.1,  # Adjust the y-coordinate for label placement
    showarrow=False,
    font=dict(color='white', size=14, family='Arial, sans-serif'),  # Set the text size and family
)
###############################################################################
#               BAttery Invented
fig.add_shape(
    type='rect',
    x0='1800-01-01',
    x1='1805-12-31',
    y0=0,
    y1=len(data_people['Name']),
    fillcolor='rgba(255, 0, 0, 0.2)',
    line=dict(color='black', width=3),
    layer='below',
)

# Add a label to the vertical bar with line breaks and bold text
fig.add_annotation(
    text='<b>Battery<br>Invented<br>',
    x='1801-06-15',  # Adjust the x-coordinate for label placement
    y=len(data_people['Name']) * -0.1,  # Adjust the y-coordinate for label placement
    showarrow=False,
    font=dict(color='white', size=14),  # Set the text size
)


##############################################################################

################################################################################

# Customize the chart appearance
fig.update_layout(
    title='Famous People Timeline with Events',
    xaxis_title='',
    yaxis_title='',
    height=1000,
    width=1200,
    plot_bgcolor='darkblue',
    paper_bgcolor='darkblue',
    font_color='white',
)

fig.show()
```
